```python
from tabletop.ee_sim_env import make_ee_sim_env
```

```python
env = make_ee_sim_env('aloha')
```

```python
ts = env.reset()
```

```python
import matplotlib.pyplot as plt
action = [-0.1, 0.15, 0.05, 0, 0, 0, 1, -0.1, 0.15, 0.05, 0, 0, 0, 1]
img_list = []
for i in range(300):
    ts = env.step(action)
    img_list.append(ts.observation['images']['front'])
    # print(ts.observation['ee_rpy_pos'])
```

```python
import cv2
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 'mp4v' for MP4 format
height, width, _ = img_list[0].shape
print(height, width)
out = cv2.VideoWriter(f'output_aloha.mp4', fourcc, 50, (width, height))
for img in img_list:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    out.write(img)
out.release()
```
